```python
# 1. if root is None then root is new node.
# 2.     back to function
# 3. append the root to QUEUE
# 4. until the QUEUE GETS EMPTY PUT WHILE
# 5. POP the element inside it and store it
# 6. if left is empty then CONNECT and BREAK
# 7.     back to function 
# 8. again step3 to step5
# 9. left is NOT EMPTY so APPEND LEFT node to QUEUE
# 10. connect right and break
# 11.     back to function
# 12. if left is full and right is full
# 13. both the node gets into the queue
# 14 then go

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        # If data is None, we skip inserting (as per level-order traversal rules)
        if data is None:
            return
        
        newnode = Node(data)
        
        if self.root is None:
            self.root = newnode
            print('r-',newnode.data)
            return
        
        queue = [] #9
        queue.append(self.root)
        
        

        while queue:
            root = queue.pop(0)

            # Insert into the left child if it's empty
            if root.left is None:
                root.left = newnode
                print('l-',newnode.data)
                break
            else:
                queue.append(root.left)
                print('ql===',root.data)

            # Insert into the right child if it's empty
            if root.right is None:
                root.right = newnode
                print(newnode.data)
                break
            else:
                queue.append(root.right)
                print('qr===',root.data)

# Driver code
tree = Tree()
root = [3, 9, 20,1,2,None,20,30]  # Level-order input
count=0
for i in root:
    count+=1
    tree.insert(i)
print('c',count)

# The expected output will print the inserted node values in the order they're inserted:
# 3
# 9
# 20
# 15
# 7
```
